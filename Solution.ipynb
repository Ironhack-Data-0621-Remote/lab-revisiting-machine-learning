```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
num = data1.select_dtypes(include=numerics)
num.head()
```

```python
cat=data1.select_dtypes(exclude=numerics)
cat.head()
```

```python
print(cat['GENDER'].value_counts())
cat['GENDER'] = cat['GENDER'].fillna('F')
```

```python
def clean_gender(x):
    if str(x).startswith('M'):
        return 'M'
    elif str(x).startswith('F'):
        return 'F'
    else:
        return 'O'

cat['GENDER'] = cat['GENDER'].apply(clean_gender)

cat['GENDER'].value_counts()
```
